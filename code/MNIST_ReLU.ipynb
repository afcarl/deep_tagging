```python
# Training set
x = data[0][0]
y = data[0][1]

# Validation set
vx = data[1][0]
vy = data[1][1]

# Test set
tx = data[2][0]
ty = data[2][1]

N = x.shape[0]
D = x.shape[1]
M = len(np.unique(y))
```

Train a network

```python
reload(relu_net)

# Number of units per hidden layer
K = 800
# Minibatch size
S = 100
# Number of hidden layers
depth = 2
# Dropout probability
delta = 0.5

rng = theano.sandbox.cuda.rng_curand.CURAND_RandomStreams(100003)

net = relu_net.relu_classifier(D, depth, K, M, rng)
last_logp = -1e100
eta = 0.5
nepochs = 30
print 'epoch\tstepsize\tE[logp(y|x)]\terror'
# An epoch is one pass over the training data
for epoch in xrange(nepochs):
    # Evaluate accuracy on held-out validation set
    predictions = net.predict(vx.T)
    logp = net.logp(vx.T, vy)
    accuracy = np.mean(np.argmax(predictions, 0) == vy)
    print '%d\t%.3e\t%.3e\t%.4f' % (epoch, eta, logp, 1-accuracy)
    sys.stdout.flush()

    # If log probability is going down, decrease step size
    if (logp < last_logp):
        eta *= 2.0 ** -0.5
    last_logp = logp

    # Execute stochastic gradient updates
    for i in xrange(0, N, S):
        inds = np.arange(i, i+S, dtype=int)
        net.grad_step(x[inds, :].T, y[inds], eta, delta)
        
# Compute test error
print 'Test log probability: %.3e' % net.logp(tx.T, ty)
predictions = net.predict(tx.T)
print 'Test error: %.3f' % (1 - np.mean(np.argmax(predictions, 0) == ty))
```
